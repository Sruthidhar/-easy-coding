```python
#function to determine the net revenue
def compute_rev(x):
    if x > 600:
        return x*0.25
    elif x > 350:
        return x*0.15
    elif x > 100:
        return x*0.10
    else:
        return x*0.05

df['Revenue'] = df['discounted_price'].apply(compute_rev)
df.head()
```

```python
# get the total revenue and print it
total_rev = df['Revenue'].sum()
print('The net revenue generated by the ECom company across all orders is', total_rev, 'rupees')
```

```python
df['BrandRevenue']= df['discounted_price']-df['Revenue'] 
df.groupby(['Brand'])['BrandRevenue'].sum().sort_values(ascending = False).head(10)
```
